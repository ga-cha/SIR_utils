```python
from brainsmash.mapgen.base import Base
import numpy as np
import csv

# load parcellated structural neuroimaging maps
atrophy = np.loadtxt("data/emp_atrophy.csv")
distance = np.loadtxt("data/sconnLen_euc.csv", delimiter=",")

# instantiate class and generate 1000 surrogates
gen = Base(atrophy, distance)
smash_maps = gen(n=1000)
```

```python
with open('results/null_smash_atrophy.csv', 'w', newline='') as f:
    wr = csv.writer(f)
    wr.writerows(smash_maps)
```
