```python
import pandas as pd
import numpy as np
from brainsmash.mapgen.base import Base
import scipy.io
```

```python
genes = pd.read_csv('./data/gene_expr_s132_indiv.csv')         # from abagen
distmat = np.loadtxt('./data/distmat_ifod.csv', delimiter=',')   # ifod2 sift2

surrogates = pd.DataFrame(np.zeros(genes.shape), columns=genes.columns)

for g in range(100):
# for g in range(genes.shape[1]):
    gene_map = genes.iloc[:, g].to_numpy()

    gen = Base(gene_map, distmat)
    surrogates[genes.columns[g]] = gen(n=1)

gene_names = genes.columns.tolist()
scipy.io.savemat('./results/gene_names.mat', mdict={'gene_names': np.array(gene_names, dtype=object)})
```

```python
gene_pval = pd.read_csv('./data/250910_gene_pair_pvalues.csv')
risk_genes = gene_pval.loc[gene_pval['pass_both'] == 1, 'risk gene'].unique().tolist()
clear_genes = gene_pval.loc[gene_pval['pass_both'] == 1, 'clearance gene'].unique().tolist()
pass_genes = list(set(risk_genes).union(set(clear_genes)))
display(len(pass_genes))
```
