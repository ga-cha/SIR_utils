```python
# import nibabel as nib
import pandas as pd
import numpy as np
from brainsmash.mapgen.base import Base
# import os
```

```python
genes = pd.read_csv('./gene_expr_s132_indiv_z.csv')         # from abagen
distmat = np.loadtxt('./distmat_ifod.csv', delimiter=',')   # ifod2 sift2

surrogates = pd.DataFrame(np.zeros((66, 1000)))             # 1 null for each of the first 1000 genes
for g in range(1000):
    gene_map = genes.iloc[:, g].to_numpy()

    gen = Base(gene_map, distmat)
    surrogates.iloc[:, g] = gen(n=1)
    
np.savetxt('./rand_genes.csv', surrogates, delimiter=',')
```
